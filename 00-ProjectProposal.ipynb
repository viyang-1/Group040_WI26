## Research Question

Is there a relationship between the COGS course drop rate at UCSD and students' evaluations of these courses - specifically course workload (study hours per week), instructor ratings, and average course grades - using CAPE (FA22 - SP23) and SET (FA23 - FA25) evaluation data?

## Background and Prior Work

It's so common for college students to choose to drop their courses within the beginning few weeks of a new quarter, that we often overlook the significance behind this behavior: what factors could actually be correlated with this complex decision - giving up a course that is planned or even anticipated for months since registration? Specifically, aside from non-academic reasons, what aspects of a course itself may be associated with a high withdrawal rate?

One previous research has been conducted on the affective reasoning of why students might choose to withdraw courses.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Among the multiple reasons for dropping from a course, "at risk of failing the course" is ranked at the top with a 39.9% probability of withdrawing. In other words, students may tend to drop courses if they feel they may get bad grades, mainly due to the difficulty, as the courses continue. 

Another research has pointed out that "assignment too heavy" could be a reason of dropping courses as well.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) This term, first of all, may be regarded as an implication of the difficulty of the course. However, this research was conducted through community colleges, and academic workload is related to job conflict (time conflict) specifically in this case. Therefore, heavy assignments also represent time consuming tasks. 

We also discovered a prior study that takes the teaching quality into account.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) By examining SRT (students rating of teaching) data, the researchers show that students can have their own evaluations of the instructors only after taking a few classes, which is then a critical factor of if the students will drop the course. Noteworthy, this factor is isolated from the probability of receiving bad grades or being assigned heavy tasks, because teaching methods or attitudes of professors do not necessarily relate to their assignments or rating scales.

Drawing on the findings of the researches above, we include three major instructional factors - average course grades, course workload (work time), and instructor ratings - in our research of their correlations to the students' dropping rate of COGS courses at UCSD, examining if they show the same correlations in our case as well.

## Hypothesis


We predict that there is likely a strong correlation between COGS course drop rate at UCSD and students' evaluations. We predict that courses with a higher workload, lower professor rating and lower grades are likely to have a higher drop rate. All these factors represent a class with a higher difficulty, which we believe will lead to a greater number of students dropping said class. 

## Data

The "ideal" dataset that we would be looking for should contain information like instructor ratings, average course grades, number of students enrolled, and a measurement of their workload. Variables like average grade and instructor ratings will be continuous, along with discrete variables such as the amount of students enrolled, and what course they are enrolled in. We will be looking at data from the 2022 Fall quarter to 2025 Fall quarter via SET and CAPE evaluations which are publically available. The data would be stored in a table grouped by the course id / course name, and for each of these we would have data containing the average grades, instructor ratings, drop rates, and course workloads.
Some datasets that already exist include SET Evals and CAPE Evals. These are stored as URLs, and are publically available to everyone. Another public student made dataset exists in Github, containing data recorded from webreg, which include the times at which students dropped or enrolled in the class.
