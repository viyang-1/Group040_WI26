## Hypothesis


We predict that there is likely a strong correlation between COGS course drop rate at UCSD and students' evaluations. We predict that courses with a higher workload, lower professor rating and lower grades are likely to have a higher drop rate. All these factors represent a class with a higher difficulty, which we believe will lead to a greater number of students dropping said class. 

## Data

The "ideal" dataset that we would be looking for should contain information like instructor ratings, average course grades, number of students enrolled, and a measurement of their workload. Variables like average grade and instructor ratings will be continuous, along with discrete variables such as the amount of students enrolled, and what course they are enrolled in. We will be looking at data from the 2022 Fall quarter to 2025 Fall quarter via SET and CAPE evaluations which are publically available. The data would be stored in a table grouped by the course id / course name, and for each of these we would have data containing the average grades, instructor ratings, drop rates, and course workloads.
Some datasets that already exist include SET Evals and CAPE Evals. These are stored as URLs, and are publically available to everyone. Another public student made dataset exists in Github, containing data recorded from webreg, which include the times at which students dropped or enrolled in the class.
